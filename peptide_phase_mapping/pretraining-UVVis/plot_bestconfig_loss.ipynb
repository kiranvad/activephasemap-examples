```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
BEST_CONFIG_DIR = "./tune/16_4_128_1.05E-03_4"
loss = np.load(BEST_CONFIG_DIR+"/loss.npy")
iters = np.arange(1, len(loss)+1)
fig, ax = plt.subplots()
ax.plot(iters, loss)
ax.set_xlabel("Number of epochs")
ax.set_ylabel("NP loss value")
plt.savefig("./np_loss.png")
plt.show()
```
