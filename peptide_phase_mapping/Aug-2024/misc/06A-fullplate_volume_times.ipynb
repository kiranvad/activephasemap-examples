```python
import numpy as np 
import pandas as pd 
from otto import Create_Directions as CD
from otto import OT2Directions as OT2D
```

```python
times, volumes = [], []
for i in range(7):
    times_i = pd.read_csv('../opentrons/times_%d.csv'%i)
    volumes_i = pd.read_csv('../opentrons/volumes_%d.csv'%i)
    times.append(times_i)
    volumes.append(volumes_i)
```

```python
times_all_iters = pd.concat(times)
volumes_all_iters = pd.concat(volumes)
```

```python
import matplotlib.pyplot as plt 

fig, ax = plt.subplots()
ax.scatter(volumes_all_iters["Z2"].to_numpy(), 
           times_all_iters["DT4"].to_numpy()
           )
plt.show()
```

```python
times_all_iters.to_csv("./times_all_iters.csv",index=False)
volumes_all_iters.to_csv("./volumes_all_iters.csv",index=False)
```
