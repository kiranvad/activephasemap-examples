```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

```python
spectrum = pd.read_csv("./spectrum_3.csv").to_numpy()
fig, ax = plt.subplots()
ax.plot(spectrum[:,0], spectrum[:,1])
plt.show()
```

```python
from scipy.interpolate import splev, splrep
wl = spectrum[:,0]
y = spectrum[:,1]
spline = splrep(wl, y)
xt = np.linspace(400, 885, 100) # upsample to 100 points
yt = splev(xt, spline)

fig, ax = plt.subplots()
ax.scatter(wl, y, s=10, color='k')
ax.plot(xt, yt)
ax.set_xlabel("wavelength (nm)")
ax.set_ylabel("cross section")
plt.show()
```

```python
np.savez("./target_nanorod.npz", y=yt, x=xt)
```
