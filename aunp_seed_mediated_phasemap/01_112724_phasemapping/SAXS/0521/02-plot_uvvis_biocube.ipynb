```python
import numpy as np 
import matplotlib.pyplot as plt 
import pandas as pd
import os, re
```

```python
files = sorted(os.listdir("./UV/"))
```

```python
fig, ax = plt.subplots()
for i in range(4):
    print(files[i])
    water = pd.read_csv("./UV/%s"%files[i],header=None)
    ax.plot(water[0], water[1])
```

```python
for i in range(int(len(files)/4-2)):
    fig, ax = plt.subplots()
    for j in range(4):
        uvvis = pd.read_csv("./UV/%s"%files[4*i+j],header=None)
        water = pd.read_csv("./UV/%s"%files[j],header=None)
        ax.plot(uvvis[0], uvvis[1]-water[1])
```

```python
for i in range(2, int(len(files)/4-2)):
    for j in range(4):
        print(files[4*i+j])
    print(10*["*"])
```
